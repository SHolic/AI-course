# CPSC-5616EL: Gradient Descent
A simple gradient descent for linear regression with the data points (1,1), (2,2), (3,3), and (3,3). The initial weights $\theta_{0}$=0, $\theta_{1}$=0, and $\alpha$=0.05.


```python
import matplotlib.pyplot as plt

# Define the data points.
data = [(1, 1), (2, 2), (3, 3)]
m = len(data)

# Start with initial guesses for theta0 and theta1 as 0.
theta0 = 0
theta1 = 0

# Define the hyperparameters
# learning rate: determines the step size at each
# iteration while moving towards a minimum of the cost function.
alpha = 0.01

# Iteration: how many times the gradient descent algorithm will be run.
iterations = 100

# Gradient Descent Algorithm
for i in range(iterations):
    # Temporary variables to hold the summation values for the gradient calculations.
    temp_theta0 = 0
    temp_theta1 = 0
    print(f"Iteration {i}: theta0 ={theta0:.4f}, theta1 = {theta1:.4f}")
    # Calculate the gradient for each data point.
    for x, y in data:
        # Sum up the partial derivatives for each data point for theta0 and theta1.
        # Please refer to the equations we seen in class
        temp_theta0 += (theta0 + theta1 * x - y) / m
        temp_theta1 += (theta0 + theta1 * x - y) * x / m

    print(f"Iteration {i}: temp_theta0 ={temp_theta0:.4f}, temp_theta1 = {temp_theta1:.4f}")
    # Update the parameters theta0 and theta1: gradient (descent).
    # These cost functions will update in each of the iterations
    theta0 = theta0 -  alpha * temp_theta0
    theta1 = theta1 - alpha * temp_theta1
# Print the updated values of theta0 and theta1 after gradient descent.
print(f"Updated values: y = {theta0:.4f} + {theta1:.4f}x")

# Extract data point values for plotting
x_values = [x[0] for x in data]
y_values = [x[1] for x in data]

# Plot the data points
plt.scatter(x_values, y_values, color='red', marker='o', label='Data Points')

# Plot the linear regression line
plt.plot(x_values, [theta0 + theta1 * x for x in x_values], label='Linear Regression', color='blue')

# Set plot labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
